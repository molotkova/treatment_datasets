### Process numerical variables

```python
# min-max standardization for numerical features
from sklearn.preprocessing import MinMaxScaler
from utils import get_numerical_features
from utils import display_dataframe_info

numerical_features = get_numerical_features(df, "data/mimic-III-sepsis/features_raw.yaml")

scaler = MinMaxScaler()

df[numerical_features] = scaler.fit_transform(df[numerical_features])

pd.options.display.max_columns = None
display_dataframe_info(df)
display(df.describe())

```

### Reorder features for output csv (sensitive, covariate, treatment, target)

### Write data to csv and yaml 

```python
df.to_csv("data/mimic-III-sepsis/data_preprocessed.csv", index=False)

from utils import get_categorical_features
from utils import get_numerical_features
from utils import get_treatment_features
from utils import get_target_features
from utils import get_covariate_features
from utils import get_sensitive_features

get_sensitive_features = get_sensitive_features(df, "data/mimic-III-sepsis/features_raw.yaml")

fairness_features_dict = {
    "sensitive": [get_sensitive_features(df, "data/mimic-III-sepsis/features_raw.yaml")],
    "covariate": [get_covariate_features(df, "data/mimic-III-sepsis/features_raw.yaml")],
    "treatment": [get_treatment_features(df, "data/mimic-III-sepsis/features_raw.yaml")],
    "target": [get_target_features(df, "data/mimic-III-sepsis/features_raw.yaml")],
}

structural_features_dict = {
    "numerical": get_numerical_features(df, "data/mimic-III-sepsis/features_raw.yaml"),
    "categorical": get_categorical_features(df, "data/mimic-III-sepsis/features_raw.yaml"),
}

from utils import create_yaml_structure

df_structure_yaml = create_yaml_structure(df, fairness_features_dict, structural_features_dict)

with open('data/mimic-III-sepsis/features_preprocessed.yaml', 'w') as file:
    yaml.dump(df_structure_yaml, file, sort_keys=False, default_flow_style=None)
```
