```python
import pandas as pd
import yaml
```

```python
df = pd.read_csv("data/diabetes/data.csv") 
```

## Dataset: Diabetes

```python
from utils import display_dataframe_info

pd.options.display.max_columns = None
display_dataframe_info(df)
display(df.describe())
```

### Most impactful treatment features

```python
from utils import get_treatment_features

treatment_features = get_treatment_features("data/diabetes/features_raw.yaml")

df_treatment = df[treatment_features]

# give percentage of 'no' values for each treatment feature
display((df_treatment == 'No').mean())

# remove treamtent features (medications) with > 98% 'No' values 
treatment_features_remove = (df_treatment == 'No').mean() > 0.98
df.drop(columns=df_treatment.columns[treatment_features_remove], inplace=True)
```

### Handle missing values & irrelevant features

```python
from utils import calculate_nan_percentage_of_grouped_features

missing_stat = calculate_nan_percentage_of_grouped_features(df, "data/diabetes/features_raw.yaml")
```

```python
print("Missing values before preprocessing:")
display(missing_stat)
```

Remove irrelevant features

```python
df.drop(
    [   "weight",  # 97% missing values
        "max_glu_serum", # 94% missing values
        "A1Cresult", # 83% missing values
        "encounter_id", # unique for each row
        "patient_nbr", # unique for each patient
        "payer_code" # 50% missing values + not useful for target prediction
    ], 
    axis=1, 
    inplace=True
)
```

Remove 3 entries where gender is 'Unknown/Invalid' 

```python
# remove samples where gender is 'unknown' (3 samples)

df = df[df["gender"] != "Unknown/Invalid"]

```

```python
missing_stat_after = calculate_nan_percentage_of_grouped_features(df, "data/diabetes/features_raw.yaml")
print("Missing values after preprocessing:")
display(missing_stat_after)
```

### Feature engineering

```python
from utils import map_icd9_category

# map icd9 codes of diagnosis to categories
for col in ['diag_1', 'diag_2', 'diag_3']:
    df[col] = df[col].apply(map_icd9_category)
```

### Reorder features for output csv (sensitive, covariate, treatment, target)

```python
from utils import order_columns

order = order_columns(df, "data/diabetes/features_raw.yaml")

df = df[order]
```

### Process categorical features

```python
# Hot-one encoding for categorical features
from utils import get_categorical_features
from utils import maintain_order_columns

# keep the original order of columns
original_order = df.columns.tolist()

categorical_features = get_categorical_features(df, "data/diabetes/features_raw.yaml")
df = pd.get_dummies(df, columns=categorical_features, dtype=int)

df = maintain_order_columns(df, original_order, categorical_features)

# remove duplicated columns (gender_Male, gender_Female), (change_Ch, change_No), (diabetesMed_Yes, diabetesMed_No)
df = df.drop(columns=["gender_Female", "change_Ch", "diabetesMed_No"])
```

### Process numerical variables

```python
# min-max standardization for numerical features
from sklearn.preprocessing import MinMaxScaler
from utils import get_numerical_features

numerical_features = get_numerical_features(df, "data/diabetes/features_raw.yaml")

scaler = MinMaxScaler()

df[numerical_features] = scaler.fit_transform(df[numerical_features])
```

### Write data to csv and yaml


```python
pd.options.display.max_columns = None
print("Dataset after preprocessing \n")
display_dataframe_info(df)
```

```python
from utils import get_categorical_features
from utils import get_numerical_features
from utils import get_treatment_features
from utils import get_target_features
from utils import get_covariate_features
from utils import get_sensitive_features
from utils import create_yaml_structure

df.to_csv("data/diabetes/data_preprocessed.csv", index=False)

sensitive_features_raw = get_sensitive_features("data/diabetes/features_raw.yaml")
sensitive_features = []
for feature in sensitive_features_raw:
    if feature in df.columns:
        sensitive_features.append(feature)
    else:
        for col in df.columns:
            if col.startswith(feature + "_"):
                sensitive_features.append(col)


covariate_features_raw = get_covariate_features("data/diabetes/features_raw.yaml")
covariate_features = []
for feature in covariate_features_raw:
    if feature in df.columns:
        covariate_features.append(feature)
    else:
        for col in df.columns:
            if col.startswith(feature + "_"):
                covariate_features.append(col)

treatment_features_raw = get_treatment_features("data/diabetes/features_raw.yaml")
treatment_features = []
for feature in treatment_features_raw:
    if feature in df.columns:
        treatment_features.append(feature)
    else:
        for col in df.columns:
            if col.startswith(feature + "_"):
                treatment_features.append(col)

target_features_raw = get_target_features("data/diabetes/features_raw.yaml")
target_features = []
for feature in target_features_raw:
    if feature in df.columns:
        target_features.append(feature)
    else:
        for col in df.columns:
            if col.startswith(feature + "_"):
                target_features.append(col)
    

numerical_features = get_numerical_features(df, "data/diabetes/features_raw.yaml")

# all columns of dataframe except numerical_features
categorical_features = list(set(df.columns) - set(numerical_features))


fairness_features_dict = {
    "sensitive": sensitive_features,
    "covariate": covariate_features,
    "treatment": treatment_features,
    "target": target_features
}

structural_features_dict = {
    "categorical": categorical_features,
    "numerical": numerical_features
}


df_structure_yaml = create_yaml_structure(df, fairness_features_dict, structural_features_dict, "diabetes")

with open('data/diabetes/features_preprocessed.yaml', 'w') as file:
    yaml.dump(df_structure_yaml, file, sort_keys=False, default_flow_style=None)
```
